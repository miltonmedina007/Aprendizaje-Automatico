A veces la informacion que recibimos viene en una escala dada por ejemplo las personas medimos entre 1.4 m a 2m ya mas alla o menos se condideran datos extremos para mostrar esto como centrados podemos tomar los valores y normalizarlo mediante es siguiente codigo.

```python
media_adultos= df.adults.mean()
df.adults.map(lambda p: p -media_adultos)
```

Nota que estos son los valores que tendria sin embargo estos no ha sido cambiado en tu dataset por lo cual vamos a hacer ese cambio con los siguientes codigos

```python
df.adults
```

```python
def remean_adults(row):
    row.adults = row.adults - media_adultos
    return row

df.apply(remean_adults, axis='columns')
```

# Leccion 4

Muy bien no te des por vencido ahora vamos a ver codigo que nos permita agrupar y ordenar
por ejemplo por tipo de hotel 

```python
df.groupby('hotel').stays_in_weekend_nights.sum()
```

Puedes obtener mucha informacion agrupada con este codigo, por ejemplo el maximo de noches en los fines de semana por cada categoria de hoteles

```python
df.groupby('hotel').stays_in_weekend_nights.max()
```

Tambien podemos agrupar en varias categorias con el siguiente codigo veamos como seria los hotels por meses

```python
agrupado=df.groupby(['hotel', 'arrival_date_month']).apply(lambda df: df.loc[df.stays_in_weekend_nights.idxmax()])
agrupado
```

Pidemos ordenar los valores con el siguiente codigo

```python
agrupado.sort_values(by='arrival_date_week_number', ascending=False)
```
